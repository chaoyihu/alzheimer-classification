```python
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt
```

```python
#path = r'D:\CY\Beihang_courses\cy_junior_2\Nervous_system_medical_engineering\project\ADNIdata\Sorted\FunImg\Sub_001'
path = r'D:\CY\Beihang_courses\cy_junior_2\Nervous_system_medical_engineering\project\ADNIdata\Sorted\T1Img\Sub_001'
img = nib.load(path + '\\Sub_001_MPRAGESENSE_20121114151828_301_Crop_1.nii')
img_uncrop = nib.load(path + '\\Sub_001_MPRAGESENSE_20121114151828_301.nii')
```

```python
print(img.header.get_data_shape())
print(img_uncrop.header.get_data_shape())
```

```python
img_data = img.get_fdata()
img_uncrop_data = img_uncrop.get_fdata()
```

```python
def show_slices(slices):
    fig,axes = plt.subplots(1,len(slices))
    for i,slice in enumerate(slices):
        axes[i].imshow(slice.T,cmap="gray",origin="lower")

slice_0 = img_data[85, :, :]
slice_1 = img_data[:, 128, :]
slice_2 = img_data[:, :, 85]
show_slices([slice_0, slice_1, slice_2])
plt.suptitle("Center slices for EPI image")  
```

```python
slice_0 = img_uncrop_data[85, :, :]
slice_1 = img_uncrop_data[:, 128, :]
slice_2 = img_uncrop_data[:, :, 128]
show_slices([slice_0, slice_1, slice_2])
plt.suptitle("Center slices for EPI image") 
```
